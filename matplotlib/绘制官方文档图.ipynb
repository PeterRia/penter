# 公众号：可视化图鉴

```python
import matplotlib
print(matplotlib.__version__) #查看Matplotlib版本
import pandas as pd
print(pd.__version__) #查看pandas版本
import numpy as np
print(np.__version__) #查看numpy版本
import matplotlib.pyplot as plt 
plt.rcParams['font.sans-serif'] = ['SimHei']  #设置中文

from matplotlib.ticker import MultipleLocator, FormatStrFormatter

from pylab import *
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
import matplotlib.image as mpimg
```

```python
X1 = list(np.random.uniform(0.5,1,20))
Y1 = list(np.random.uniform(1,3.3,20))

X2 = list(np.random.uniform(1,2,35))
Y2 = list(np.random.uniform(0.6,3,35))

X3 = list(np.random.uniform(2,2.5,10))
Y3 = list(np.random.uniform(0.7,2.2,10))

X5 = list(np.random.uniform(2.5,3,10))
Y5 = list(np.random.uniform(1,2,10))

X4 = list(np.random.uniform(3,3.3,10))
Y4 = list(np.random.uniform(1.3,2,10))

X = X1 + X2 + X3 + X4 + X5

Y = Y1 + Y2 + Y3 + Y4 + Y5

A = X
B = Y
```

```python
plt.rc('font',family='Avenir') 
plt.figure(figsize = (8,7), dpi=100)
plt.title("Anacomy of a figure",fontsize = 20,fontweight='heavy')

ax = plt.gca()

ax.set_xlim(0,4)
ax.set_ylim(0,4)


xmajorLocator   = MultipleLocator(1)
ax.xaxis.set_major_locator(xmajorLocator)

ymajorLocator   = MultipleLocator(1)
ax.yaxis.set_major_locator(ymajorLocator)

xminorLocator   = MultipleLocator(0.25)
ax.xaxis.set_minor_locator(xminorLocator)

yminorLocator   = MultipleLocator(0.25)
ax.yaxis.set_minor_locator(yminorLocator)


xminorFormatter = FormatStrFormatter('%0.2f')
ax.xaxis.set_minor_formatter(xminorFormatter)

tick_params(which='major',length=8,labelsize=10)
tick_params(which='minor',length=4)


ax.xaxis.grid(True, which='major',linestyle = (0,(8,4))) 
ax.yaxis.grid(True, which='major',linestyle = (0,(8,4))) 

ax.set_xlabel("X axis label",fontsize = 12)
ax.set_ylabel("Y axis label",fontsize = 12)

x = list(np.arange(0.5,3.5,0.1))
y2 = [0.1545*i**3-0.799*i**2+0.4316*i + 3.744 for i in x]
ax.plot(x,y2,color = 'blue',alpha = 0.7, linewidth=2.3,label = "Blue signal")

x = list(np.arange(0.5,3.3,0.1))
y1 = [-0.04658*i**3+0.5494*i**2-1.3151*i + 1.415 for i in x]
ax.plot(x,y1,color = 'red',alpha = 0.7, linewidth=2.3,label = "Red signal")


ax.scatter(A, B,marker = 'o',color="w",edgecolors='black')

ax.legend()


ax.text(2.9,2.65,"Gird",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(3.45,3.45,"Legend",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')
ax.text(1.5,3.85,"Title",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(-0.25,3.65,"Major tick",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')
ax.text(-0.25,3.25,"Minor tick",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')
ax.text(-0.45,2.75,"Major tick label",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')
ax.text(0.12,-0.5,"Minor tick label",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')
ax.text(-0.6,1.5,"Y axis label",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')
ax.text(-0.4,0.5,"Figure",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(0.25,0.25,"Axes",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(1.05,0.12,"    Line\n(line plot)",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(1.25,2.68,"    Line\n(line plot)",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(1.5,-0.7,"X axis label",family = "DejaVu Sans",fontsize = 11,color = 'blue',weight='heavy')
ax.text(3.25,0.5,"Splines",family = "DejaVu Sans",fontsize = 10,color = 'blue',weight='heavy')


ax.arrow(3.3,0.41,-0.15,-0.33,ec='blue',head_width=0.04)
ax.arrow(3.6,0.41,0.32,-0.15,ec='blue',head_width=0.04)


arr_lena = mpimg.imread('WechatIMG2278.png')
imagebox = OffsetImage(arr_lena, zoom=0.4)

a1 = AnnotationBbox(imagebox, (0, 4), frameon = False)
a2 = AnnotationBbox(imagebox, (3.6, 3.77), frameon = False)
a3 = AnnotationBbox(imagebox, (3, 3), frameon = False)
a4 = AnnotationBbox(imagebox, (1.3, 0.6), frameon = False)

ax.add_artist(a1)
ax.add_artist(a2)
ax.add_artist(a3)
ax.add_artist(a4)

plt.show()
```
